```python
# 웹스크래핑

# 이번엔 한국거래소에서 제공하는 상장법인 목록을 다운로드하고 종목코드를 구한 뒤
# 그걸 기준으로 네이버 금융에서 일자별 주가 데이터를 스크래핑 할 것이다.
# 그리고 그걸 캔들 차트로 그려본다.

import pandas as pd

krx_list = pd.read_html('상장법인목록.xls')
krx_list[0]
```

```python
# 종목코드가 0이 빠진 부분에 대해 map()함수로 보정해준다

# 아래 함수가 2회 실행부터는 오류가 나더라 흠
# krx_list[0].종목코드 = krx_list[0].종목코드.map('{:06d}'.format)

krx_list[0]
```

```python
# read_html()은 로컬에서 파일을 읽어올 수 있을 뿐 아니라 웹에서 바로 읽기도 가능

# 웹에서 바로 읽어오기 + 종목코드 오름차순 정렬을 해보자

#df = pd.read_html('https://kind.krx.co.kr/corpgeneral/corpList.do?method=loadInitPage#')[0]
df = pd.read_html('상장법인목록.xls')[0] # 뒤에 [0]을 달아줌으로써 리턴값을 dataframe이 되도록 조절한 것

df['종목코드'] = df['종목코드'].map('{:06d}'.format)
df = df.sort_values(by='종목코드', ascending=True)
df
```
