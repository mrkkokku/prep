```python
# 하나의 그래프 영역에 여러개의 곡선을 나타내는 방법

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)
print(x.shape) # n-dimension은 아닌걸 확인

# 아래와 같이 순서대로 x값, y값, 옵션 순으로
# 3회 반복해서 3가지 곡선이 한번에 그려지는 것임

plt.plot(x, x, 'r--', x, x**2, 'bo', x, x**3, 'g-.')
plt.show()
```

```python
# 스타일 지정하기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

# 아래와 같이 함수를 여러번 호출하여
# 표현할 곡선들을 누적할 수 있고, 또한 옵션도 각각 설정가능하다.

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)
plt.show()
```
