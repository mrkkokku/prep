```python
# xscale(), yscale() 함수를 사용해서 그래프의 축 스케일링을 할 수 있다

# 축은 기본적으로 linear 스케일로 표기되지만

# log, symlob, logit 으로 변경가능하다.

# symlog : Symmetrical Log Scale
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = x ** 3

plt.plot(x, y)

plt.xscale('symlog')
# 옵션지정하는 것이 도메인 값을 직접 넣는것에 유의

# symlog 라서
# 축의 원점을 기준으로 양, 음의 방향이 대칭적인 로그 스케일로 표시됩니다.

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# y축 스케일 지정하기

x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x, y)
# plt.yscale('linear') # 이거면 평범하게 exponential 나왔겠지만
plt.yscale('log')      # y만 log스케일로 지정

# 로그 스케일은 지수함수 (Exponential function)와 같이 
# 기하급수적으로 변화하는 데이터를 표현하기에 적합합니다.

plt.show()
```
