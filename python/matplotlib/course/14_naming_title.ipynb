```python
# pyplot 모듈 내의 title() 함수를 이용해 그래프의 타이틀을 설정할 수 있음

# 타이틀을 표시하고, 위치를 조절하고, 폰트와 스타일을 조절하는 방법을 알아보자

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.title('Graph Title') # 이렇게 타이틀을 지정할 수 있다.

plt.show()
```

```python
# 위치와 오프셋 지정하기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)

# 아래와 같이 타이틀 이름 외에 loc, pad 옵션을 지정하여 타이틀에 대한
# 스타일을 지정할 수 있다

# {‘left’, ‘center’, ‘right’} 중 선택할 수 있으며 디폴트는 ‘center’입니다.
# pad 파라미터는 타이틀과 그래프와의 간격을 포인트 단위로 설정합니다.

plt.title('Graph Title', loc='right', pad=20)

plt.show()
```

```python
# 폰트 지정하기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.title('Graph Title', loc='right', pad=20)

# 이전에 배운대로
# 딕셔너리안에 폰트정보를 담아두고
# 이것을 그대로 활용가능하다

# ‘fontsize’는 포인트 단위의 숫자를 입력하거나
# ‘smaller’, ‘x-large’ 등의 상대적인 설정을 할 수 있습니다.
# ‘fontweight’에는
# {‘normal’, ‘bold’, ‘heavy’, ‘light’, ‘ultrabold’, ‘ultralight’}
# 와 같이 설정할 수 있습니다.

title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
plt.title('Graph Title', fontdict=title_font, loc='left', pad=20)

plt.show()
```

```python
# 타이틀 얻기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
title_right = plt.title('Graph Title', loc='right', pad=20)

title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
title_left = plt.title('Graph Title', fontdict=title_font, loc='left', pad=20)

# plt.title() 함수는 타이틀을 나타내는 Matplotlib text 객체를 반환합니다.
# get_position()과 get_text() 메서드를 사용해서 텍스트 위치와 문자열을 얻을 수 있습니다.

print(title_left.get_position())
print(title_left.get_text())

print(title_right.get_position())
print(title_right.get_text())

plt.show()
```
