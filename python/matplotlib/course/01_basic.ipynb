```python
import matplotlib.pyplot as plt

# https://wikidocs.net/book/5011

# matplotlib.pyplot 모듈은 MATLAB과 비슷하게 명령어 스타일로 동작하는 함수의 모음
# matplotlib.pyplot은 각각의 함수를 사용해서 간편하게 그래프를 만들고 변화를 줄 수 있다.


# 기본 그래프 그리기
# plot() 함수는 리스트의 값들이 y값이라 가정하고, x값 [0, 1, 2, 3]을 자동으로 만든다
plt.plot([1, 2, 3, 4])
plt.show()
```

```python
import matplotlib.pyplot as plt

# plot() 함수는 여러개의 인자를 받을 수 있다

plt.plot([1, 2, 3, 4], [1, 4, 9, 4])
plt.show()
```

```python
import matplotlib.pyplot as plt

# plot()의 3번째 인자값으로 선의 색상과 형태를 지정할 수 있는데
# ro는 Red-O 표시이다, b- 는 Blue-Line 의 의미 등등
# axis()함수는 축의 범위를 뜻한다
# [x min, x-max, y-min, y-max]

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 여러개의 그래프 그리기


# 200ms 간격으로 균일하게 샘플된 시간이라 할 때
t = np.arange(0., 5., 0.2)

# 빨간 대시, 파란 사각형, 녹색 삼각형
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
