```python
import matplotlib.pyplot as plt

# plot()에 리스트를 입력하면, y값으로 인식한다.
# 그리고 그 출력은 꺽은선 그래프로 나오게 됨
plt.plot([2, 3, 5, 10])
plt.show()

# 이것의 결과는 (0,2) (1,3) (2,5) (3,10)을 잇는 꺽은선 그래프다
```

```python
import matplotlib.pyplot as plt

# x, y값 둘 다 입력하기

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.show()

# 놀랍게도, 이것의 결과는 (1,2) (2,3) (3,5) (4,10)을 잇는 꺽은선 그래프가 나온다
```

```python
import matplotlib.pyplot as plt

# 레이블이 있는 데이터 사용하기
# 딕셔너리 안에 키와 밸류로 값을 넣어놓고
# plot() 안에 키 값 두개 넣고, data는 따로 전달하는 모습 확인

data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}

plt.plot('data_x', 'data_y', data=data_dict)
plt.show()
```
