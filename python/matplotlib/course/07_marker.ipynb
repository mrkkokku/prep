```python
# 특별한 설정이 없으면 그래프가 실선으로 그려지지만,
# 마커를 활용해 마커형태의 그래프를 그릴 수 있다

import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'bo') # bo 자체가 마커옵션 중 하나이고 Blue-O를 의미
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

# 선과 마커를 동시에 나타낼 수 있는데,
# bo는 마커, -는 solid line을 의미하므로 아래와 같고
# --역시 마커와 dash line을 의미하므로 그림을 참고하자

# plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'bo-')    # 파란색 + 마커 + 실선
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'bo--')     # 파란색 + 마커 + 점선
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

# plot() 함수의 marker 파라미터를 사용하면 더욱 다양한 마커형태를 지정 가능
# 지정가능한 옵션들에 대해선 러퍼런스 사이트를 찾아가서 참조하자

plt.plot([4, 5, 6], marker="H")
plt.plot([3, 4, 5], marker="d")
plt.plot([2, 3, 4], marker="x")
plt.plot([1, 2, 3], marker=11)
plt.plot([0, 1, 2], marker='$Z$')
plt.show()
```
