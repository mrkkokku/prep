```python
import matplotlib.pyplot as plt

# 마치 마커지정하듯이 동일하게 색상을 뜻하는 알파벳으로 색깔을 지정할 수 있다
# 이것이 format string을 활용한 방법임

plt.plot([1, 2, 3, 4], [2.0, 3.0, 5.0, 10.0], 'r')
plt.plot([1, 2, 3, 4], [2.0, 2.8, 4.3, 6.5], 'g')
plt.plot([1, 2, 3, 4], [2.0, 2.5, 3.3, 4.5], 'b')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```

```python
import matplotlib.pyplot as plt

# 아래는 위와 다르게, 컬러 키워드를 사용하여 색상을 지정하는 방식이다.
# 좀 더 상세하게 지정하려면 이 방법이 좋을 것 같다

plt.plot([1, 2, 3, 4], [2.0, 3.0, 5.0, 10.0], color='limegreen')
plt.plot([1, 2, 3, 4], [2.0, 2.8, 4.3, 6.5], color='violet')
plt.plot([1, 2, 3, 4], [2.0, 2.5, 3.3, 4.5], color='dodgerblue')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```

```python
import matplotlib.pyplot as plt

# 유사하게 Hex Code를 활용해서 더욱 더 다양한 색상을 지정할 수 있다.
# 게다가 이번엔 marker, linestyle까지 한꺼번에 지정하는 것도 해보자.
# 파이썬의 특징으로 오버로딩할때 파라미터를 받을 옵션명을 
# 지정 해 주면 인풋값 순서 관계없이 모두 입력이 가능하다는 것도 확인

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], color='#e35f62',
        marker='o', linestyle='--')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```
