```python
import matplotlib.pyplot as plt

# Legend = 범례
# legend() 함수를 이용해서 그래프에 범례를 표기할 수 있다.

# 그래프 영역안에 범례를 나타내기 위해선 plot()함수 안에
# label 문자열 속성을 지정한 후에 legend() 함수까지 콜 해줘야 한다
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price (%)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# legend 위치 지정하기

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price (%)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
#plt.legend(loc=(0.0, 0.0)) # 그래프의 왼쪽아래에 위치 시키기
plt.legend(loc=(1.0, 1.0))  # 그래프의 오른쪽 위 끝에 위치시키기
plt.show()
```

```python
import matplotlib.pyplot as plt

# 위치를 정량적이 아닌, 정성적으로 위치 시킬 수 있다

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='lower right') # 이렇게도 사용 가능

plt.show()
```

```python
import matplotlib.pyplot as plt

# 열 개수 지정하기
# legend 함수의 ncol 파라미터는 범례에 표시될 텍스트의 열의 개수를 지정
# ncol=1 이면, 범례에 price와 demand가 2행 1열로 표기되며
# ncol=2 이면, 범례에 1행 2열로 표기 된다.

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='best', ncol=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 폰트크기 지정하기

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc='best') # 이렇게도 사용가능 
plt.legend(loc='best', ncol=2, fontsize=14) # 폰트크기까지 지정함

plt.show()
```

```python
import matplotlib.pyplot as plt

# frameon 파라미터는 범례 텍스트 상자의 테두리를 표시할지 여부를 설정
# frameon = False로 하면 테두리가 표기되지 않음
# shadow 속성을 사용해, 텍스트 상자에 그림자효과를 줄 수 있음

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc='best')
plt.legend(loc='best', ncol=2, fontsize=14, frameon=True, shadow=True) # 다양한 속성을 지정함

plt.show()
```
