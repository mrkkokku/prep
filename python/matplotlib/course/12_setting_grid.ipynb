```python
# data의 위치를 더 명확하게 나타내기위해
# 그래프에 그리드를 표기할 수 있다.

# pyplot모듈 내 grid함수를 이용해 이걸 해보자

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2) # array를 만들고

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.grid(True) # grid함수에 True 넣음

plt.show()
```

```python
# 축 지정하기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)
plt.grid(True, axis='y')  # 위와 같지만 axis='y'옵션 추가함
                          # 이렇게하면 y축에 대해서만 grid 표시함
                          # (마치 axis=1과 같은 느낌으로)
                          # 설정가능한 옵션은
                          # both, x, y중 하나만 선택가능 하고
                          # 기본값은 both이다.

plt.show()
```

```python
# 스타일 설정하기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)

# 아래와 같이 그리드에 대해 color, alpha, linestyle 등 여러
# 옵션을 직접 지정가능
# 특히 which 파라미터를 이용해 major, minor, both등을 사용하면
# 주눈금, 보조눈금에 각각 그리드를 표시할 수 있습니다.

plt.grid(True, axis='y', color='red', alpha=0.5, linestyle='--')

plt.show()
```
