```python
'''
The relationship between gravitational force and
distance between two bodies.
'''

# Draw the graph
def draw_graph(x,  y):
    plt.plot(x, y, marker='o')
    plt.xlabel("Distance in meters")
    plt.ylabel("Gravitational force in newtons")
    plt.title("Gravitational force and distance")
    plt.show()

def generate_F_r():
    # Generete values for r
    r = range(100, 1001, 50)

    # Empty list to store the calculated values of F
    F = []

    # Constant, G
    G = 6.674 * (10 ** -11)
    
    # Two masses
    m1 = 0.5
    m2 = 1.5

    # Calculate force and add it to the list, F
    for dist in r:
        force = G*(m1*m2) / (dist**2)
        F.append(force)

    # Call the draw_graph function
    draw_graph(r, F)

generate_F_r()
```

```python
# Projectile Motion
import math
'''
Generate equally spaced floating point numbers
between two given values
'''

def frange(start, final, increment):
    numbers = []

    while start < final:
        numbers.append(start)
        start = start + increment
    
    return numbers


# Draw the trajectory of a body in projectil motion

def draw_graph(x, y):
    plt.plot(x, y)
    plt.xlabel("x-coordinate")
    plt.ylabel("y-coordinate")
    plt.title("Projectile motion of a ball")


def draw_trajectory(u, theta):
    theta = math.radians(theta)
    g = 9.8

    # Time of flight
    t_flight = 2 * u * math.sin(theta) / g

    # Find time intervals
    intervals = frange(0, t_flight, 0.001)

    # List of x and y coordinates
    x = []
    y = []

    for t in intervals:
        x.append(u * math.cos(theta) * t)
        y.append(u * math.sin(theta) * t - 0.5 * g * t**2)
    
    draw_graph(x, y)
'''
try:
    u = float(input("Enter the initial velocity (m/s): "))
    theta = float(input("Enter the angle of projection (degrees): "))
except ValueError:
    print("You entered an invalid input")
else:
    draw_trajectory(u, theta)
    plt.show()
'''

draw_trajectory(25, 60)
plt.show()
```

```python
# Comparing the Trajectory at Different Initial Velocities


# List of three different initial velocities
u_list = [20, 40, 60]
theta = 45

for u in u_list:
    draw_trajectory(u, theta)

# Add a legend and show the graph
plt.legend(['20', '40', '60'])
plt.show()

# 이건 실제로 plt.plot이 3회 수행된 결과임
```
