```python
# 결측값이 있다면
file_cctv.fillna(0) # NaN을 0으로 채운다

file_cctv.dropna() # NaN이 포함된 행을 삭제한다
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))
plt.title("각 자치구별 CCTV 개수", fontsize=16)
plt.plot(file_cctv['구 분'], file_cctv['CCTV 총계'])

plt.show()
```

```python
plt.figure(figsize = (12, 4))
plt.title('각 자치구별 쓰레기무단투기용 CCTV 개수', fontsize = 16)
plt.bar(file_cctv['구 분'], file_cctv['CCTV 총계'], color = 'green')

plt.show()
```

```python
plt.title('각 기관별 CCTV 총계', fontsize = 16)
plt.boxplot(file_cctv['CCTV 총계'], labels = ['CCTV'])
```

```python
# 여기부터는 데이터 분석

import numpy as np

avg = np.mean(file_cctv['CCTV 총계']) # 평균 구하기

print(avg)
```

```python
median = np.median(file_cctv['CCTV 총계'])

print(f"중앙값 : {median}")
print(f"최소값 : {np.min(file_cctv['CCTV 총계'])}")
print(f"최대값 : {np.max(file_cctv['CCTV 총계'])}")
```

```python
# 데이터 통계 요약하기

print(file_cctv.describe()) # 수치연산 불가한 열은 제외됨
```

```python
# histogram을 이용해 최빈값 구하기

plt.title("각 기관별 CCTV 총계", fontsize=16)
plt.hist(file_cctv['CCTV 총계'], bins=50) # bins는 나누는 영역의 개수

plt.show()

# 3000중반 쪽에 몰려있음을 확인
```
