```python
# 동적사이트에선 전체 소스가 다운로드 되지않기 때문에 selenium의 도움을 받아야한다.

from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome('D:/chromedriver-win64/chromedriver.exe')

# 브라이튼 최근 경기 결과 크롤링 해보기
url = "https://www.livesport.com/kr/team/brighton/2XrRecc3/"

driver.implicitly_wait(3) # 페이지 로드까지 기다려주는 시간 (s)
driver.get(url)           # 페이지 가져옴
```

```python
page = driver.page_source # 페이지를 저장

bs_obj = BeautifulSoup(page, "html.parser")

print(bs_obj)
```

```python
result = bs_obj.find_all("span", {"class":"wcl-simpleText_Asp-0 wcl-scores-simpleText-01_pV2Wk"})

wcnt = 0
tcnt = 0
lcnt = 0
for i in range(len(result)):
    if result[i].text == '승':
        wcnt += 1
    elif result[i].text == '무':
        tcnt += 1
    elif result[i].text == '패':
        lcnt += 1

report = {"win":wcnt, "draw":tcnt, "lose":lcnt}
print(report)
```
