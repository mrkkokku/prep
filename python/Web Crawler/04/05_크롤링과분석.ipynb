```python
from bs4 import BeautifulSoup  
from urllib.request import urlopen
import serviceKey as sKey

import pandas as pd

# 아파트매매 실거래 (국토교통부)
endpoint = 'https://apis.data.go.kr/1613000/RTMSDataSvcAptTrade/getRTMSDataSvcAptTrade?'
serviceKey = sKey.serviceKey()
numOfRows = '9999'
pageNo = '1'

lawdCd = '31110'
dealYmd = '202507'

url = endpoint + 'serviceKey=' + serviceKey + '&numOfRows=' + numOfRows + '&pageNo=' + pageNo + '&LAWD_CD=' + lawdCd + '&DEAL_YMD=' + dealYmd

html = urlopen(url)
bs_obj = BeautifulSoup(html, "html.parser")

print(url)
print(bs_obj)
```

```python
item = bs_obj.find_all("item")

print(item)
```

```python
umdnm = bs_obj.find_all("umdnm")

print(umdnm)

for idx, elem in enumerate(umdnm):
    print(idx, elem.text)
```

```python
aptnm = bs_obj.find_all('aptnm')

print(aptnm)

for idx, elem in enumerate(aptnm):
    print(idx, elem.text)
```

```python
dealamount = bs_obj.find_all('dealamount')

print(dealamount)

for idx, elem in enumerate(dealamount):
    print(idx, elem.text)
```

```python
dealday = bs_obj.find_all('dealday')

print(dealday)

for idx, elem in enumerate(dealday):
    print(idx, elem.text)
```

```python
umdNmList = []
aptNmList = []
dealAmountList = []
dealDayList = []

for idx, elem in enumerate(umdnm):
    umdNmList.append(elem.text)

for idx, elem in enumerate(aptnm):
    aptNmList.append(elem.text)

for idx, elem in enumerate(dealamount):
    dealAmountList.append(int(elem.text.replace(',','')))

for idx, elem in enumerate(dealday):
    dealDayList.append(int(elem.text))


print(umdNmList)
print(aptNmList)
print(dealAmountList)
print(dealDayList)
```

```python
data = {
    '동': umdNmList,
    '아파트명': aptNmList,
    '거래가격': dealAmountList,
    '거래일': dealDayList
}

df = pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt

df = df.sort_values(by='거래일')

plt.figure(figsize=(20, 10))
plt.plot(df['거래일'], df['거래가격'], marker='o')
plt.show()
```
